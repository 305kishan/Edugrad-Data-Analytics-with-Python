B) What does ConvertedSalary column store?

```python
# C) Write code to Verify whether the ConvertedSalary column contains any null value or not

df_survey['ConvertedSalary'].isnull().any()

# Yes it contains NUll values
```

D) Based on the output of part C), remove the rows containing Null values in a separate Dataframe called df_sal_NonNa

```python
df_sal_NonNa = df_survey.dropna(axis=0, subset=['ConvertedSalary'])
```

```python
df_sal_NonNa.head()
```

E). Use seaborn library to plot histogram for non-null values of ConvertedSalary column


```python
sns.distplot(df_sal_NonNa['ConvertedSalary'],kde = False,bins =30)
sns.distplot(df_sal_NonNa['ConvertedSalary'],kde = False,bins =500)
plt.show()

#This plot contains Plot histogram for df_sal_NonNa having bin size as 30 and 500
# Blue one is with binsize 500 and brown is for binsize 30
```

F). Create a new Dataframe variable called df_sal_low, to store non-null ConvertedSalary whose value is less than or equal to 250000.


```python
df_sal_low = df_sal_NonNa[df_sal_NonNa['ConvertedSalary'] <= 250000]
```

```python
sns.distplot(df_sal_low['ConvertedSalary'], kde=False)

#this plot shows the salary of all developers who have their converted less than or equal to 250000
```

G) Plot Salary distribution for both df_sal_NonNa and df_sal_low using a KDE Plot

```python
sns.distplot(df_sal_NonNa['ConvertedSalary'],kde = True,bins = 30)
sns.distplot(df_sal_low['ConvertedSalary'], kde=True,bins = 30)


#kernel density estimation is a non-parametric way to estimate the probability density function of a random variable.
#this graph shows a kde plot for df_sal_NonNa( No empty value in ConvertedSalary) and df_sal_low( ConvertedSalary less than or equal to 25000)
```
