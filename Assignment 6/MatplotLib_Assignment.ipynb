```python
x = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
Popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

x_pos = np.arange(len(x))
```

```python
# PART _ 1
```

```python
x_pos = np.arange(len(x))
plt.bar(x_pos, Popularity, color='blue')
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.title("PopularitY of Programming Language")
plt.xticks(x_pos, x)

plt.grid(which='major', linestyle='-', linewidth='1', color='red')

plt.show()
```

```python
# PART _ 2
```

```python
colors = ['b','g','m','k','y','c']
```

```python
plt.barh(x_pos, Popularity, color=colors)
plt.ylabel("Languages")
plt.xlabel("Popularity")
plt.title("PopularitY of Programming Language")
plt.yticks(x_pos, x)

plt.grid(which='major', linestyle='-', linewidth='1', color='red')

plt.show()
```

```python
fig, ax = plt.subplots()

ax.add_patch(plt.Circle((4, 0), 1.5, color='y'))
plt.plot(4, 0, 'go')
ax.annotate('(4,0)',xy=(4,0),xytext=(4,-2),arrowprops=dict(color='m', shrink =0.1))

ax.add_patch(plt.Circle((0, 0), 2, hatch = '/'))
plt.plot(0, 0, 'go')
ax.annotate('(0,0)',xy=(0,0),xytext=(-2,-2),arrowprops=dict(color='m', shrink =0.1))


ax.set_aspect('equal', adjustable='datalim')
ax.plot()  
plt.show()
```
